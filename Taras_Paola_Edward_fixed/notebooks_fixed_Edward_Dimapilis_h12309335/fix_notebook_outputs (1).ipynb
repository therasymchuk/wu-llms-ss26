# Fix Notebook Outputs for GitHub

Some notebooks save interactive widget outputs (like tqdm progress bars) that GitHub cannot render. This notebook strips those outputs so the files display correctly on GitHub.

**What it does:**
- Removes widget metadata from the notebook header
- Clears all cell outputs and execution counts
- Saves a cleaned copy without touching the original

**How to use:**
1. Mount your Google Drive
2. Update the file paths in the configuration cell
3. Run all cells

## Step 3: Configure your notebook paths

Update the paths below to match your Google Drive folder structure.
Each entry is a pair: `(input_path, output_path)`.

Add or remove entries to match how many notebooks you want to clean.

```python
# Update these paths before running
notebooks = [
    (
        "/content/drive/MyDrive/wu-llms-ss26/01_inference.ipynb",
        "/content/drive/MyDrive/wu-llms-ss26/01_inference_fixed.ipynb"
    ),
    (
        "/content/drive/MyDrive/wu-llms-ss26/03_rag.ipynb",
        "/content/drive/MyDrive/wu-llms-ss26/03_rag_fixed.ipynb"
    ),
    (
        "/content/drive/MyDrive/wu-llms-ss26/Austrian_Tax_Law_LLM_Phase2.ipynb",
        "/content/drive/MyDrive/wu-llms-ss26/Austrian_Tax_Law_LLM_Phase2_fixed.ipynb"
    ),
    (
        "/content/drive/MyDrive/wu-llms-ss26/evaluation_local.ipynb",
        "/content/drive/MyDrive/wu-llms-ss26/evaluation_local_fixed.ipynb"
    ),
]
```

## Step 4: Clean the notebooks

```python
import nbformat

for input_path, output_path in notebooks:
    # Load the notebook
    with open(input_path, "r") as f:
        nb = nbformat.read(f, as_version=4)

    # Remove widget metadata that causes GitHub rendering errors
    if "widgets" in nb.metadata:
        del nb.metadata["widgets"]

    # Clear all cell outputs and reset execution counts
    for cell in nb.cells:
        if cell.cell_type == "code":
            cell.outputs = []
            cell.execution_count = None

    # Save the cleaned notebook
    with open(output_path, "w") as f:
        nbformat.write(nb, f)

    print(f"Cleaned: {input_path.split('/')[-1]}  -->  {output_path.split('/')[-1]}")

print("\nAll done.")
```

## Step 5: Verify the output files exist

```python
import os

for _, output_path in notebooks:
    exists = os.path.exists(output_path)
    status = "OK" if exists else "NOT FOUND"
    print(f"[{status}] {output_path.split('/')[-1]}")
```
